```python
from collections import Counter
import json
import numpy as np
plt.rcParams["font.family"] = "Times New Roman"

figure(num=None, figsize=(15, 6))

with open("../conference_medium_data/new_data/author_info_final_paper.json") as f:
    author_data = json.load(f)
num_posts = []
for key in author_data:
    num_posts.append(len(key['other_posts']))
# sorted(num_posts, reverse=True)

post_freq = Counter([item for item in num_posts])

import operator
sorted_post_freq = sorted(post_freq.items(), key=operator.itemgetter(0))

n_0_10 = 0
n_11_20 = 0
n_21_30 = 0
n_31_40 = 0
n_40_ = 0
for pair in sorted_post_freq:
    if pair[0] <= 10:
        n_0_10 += pair[1]
    elif pair[0] > 10 and pair[0] <= 20:
        n_11_20 += pair[1]
    elif pair[0] > 20 and pair[0] <= 30:
        n_21_30 += pair[1]
    elif pair[0] > 30 and pair[0] <= 40:
        n_31_40 += pair[1]
    else:
        n_40_ += pair[1]


font = {'family' : 'Times New Roman',
        'weight' : 'bold',
        'size'   : 18}

plt1.rc('font', **font)
#plt.rcParams["font.family"] = "Times New Roman"


labels = '1-10 posts', '11-20 posts', '21-30 posts', '31-40 posts', 'More than 41 posts'
sizes = [n_0_10, n_11_20, n_21_30, n_31_40, n_40_]

#Labels=['0', '1', '2', '3', '4', '5', '6', '7 or more']
y_pos=np.arange(len(labels))
plt1.bar(y_pos + 0, sizes, width=0.5, color = 'navy', label='Number of posts')
#plt.bar(y_pos + 0.25,CommentCount_for_Bar, width=0.25,color = 'c',label = 'Comments')
plt1.xticks(y_pos, labels)
plt1.legend(loc='best')
plt1.ylabel('Number of Authors')
plt1.xlabel('Number of posts')
#plt1.title("Number of posts based on the response count")
plt1.savefig('medium_author_post_count.pdf', dpi=300)
```

```python
import pandas as pd
import csv
import time
import numpy as np
from matplotlib import pyplot as plt
from numpy import int32
from datetime import datetime
import operator
import json
from matplotlib.pyplot import figure
font = {'family' : 'Times New Roman',
        'weight' : 'bold',
        'size'   : 13}

plt.rc('font', **font)

#data_sheet_Questions = pd.read_json('final_data_removing_duplicacy.json', encoding='latin-1')
#Question_count = data_sheet_Questions.shape[0]

# print("No of Posts in Medium = "+format(Question_count))
# print(data_sheet_Questions.CreationDate)
# print(data_sheet.dtypes)
# data_sheet_Questions['publish_date'] = pd.to_datetime(data_sheet_Questions['publish_date'])
# print(data_sheet_Questions.publish_date)
#figure(num=None, figsize=(12, 9))

with open("../conference_medium_data/new_data/related_data_rm_duplicacy_conf_final_2.json","r") as f:
    data = json.load(f)


def dates_4_posts(tag1, tag2):
    count19 = 0
    count18 = 0
    count17 = 0
    count16 = 0
    count15 = 0
    count14 = 0

    #l = data_sheet.sort_values(by='CreationDate')

    month2019 = [0] * 13
    month2018 = [0] * 13
    month2017 = [0] * 13
    month2016 = [0] * 13
    month2015 = [0] * 13
    month2014 = [0] * 13
    
    
    for key in data:
        if tag1 in key['content'].lower() or tag1 in key['title'].lower() or\
                        tag2 in key['content'].lower() or tag2 in key['title'].lower():
            date = str(key["publish_date"])
    #for k in data_sheet_Questions.publish_date:
     #   date, time = str(k).split(" ")  # Date and time splitting
        # date parsing
            y, m, d = date.split("-")
            if y == "2019":
                m = int(m)
                month2019[m] += 1
                count19 += 1

            elif y == "2018":
                m = int(m)
                month2018[m] += 1
                count18 += 1

            elif y == "2017":
                m = int(m)
                month2017[m] += 1
                count17 += 1

            elif y == "2016":
                m = int(m)
                month2016[m] += 1
                count16 += 1

            elif y == "2015":
                m = int(m)
                month2015[m] += 1
                count15 += 1

            elif y == "2014":
                m = int(m)
                month2014[m] += 1
                count14 += 1
    return month2016[1:] + month2017[1:] + month2018[1:] + month2019[1:9]


sec_vul = dates_4_posts("security", "vulnerability")
reentrancy = dates_4_posts("reentrancy", "re-entrancy")

Months=['January 16','','','','','','','','','','','',
        'January 17','','','','','','','','','','','',
        'January 18','','','','','','','','','','','',
        'January 19','','', '', '', '', '', 'August 19']


y_pos=np.arange(len(Months))

import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"

df = pd.DataFrame({"Months": Months,"Reentrancy_List":reentrancy, "Sec_Vul":sec_vul})

fig = plt.figure()

ax = df['Sec_Vul'].plot(kind="bar", alpha=0.7)
ax.set_xlabel('January 2016 - August 2019', size = 13)

ax2 = ax.twinx()
ax2.plot(ax.get_xticks(),df['Reentrancy_List'],marker='o', c='navy', linewidth=4)

ax.set_xticklabels(df['Months'],rotation = 0)
ax.set_ylim(0,1.3*df["Sec_Vul"].max())
ax2.set_ylim(0,1.3*df["Reentrancy_List"].max())
ax.set_ylabel('New Questions per Month Mentioning "Security/Vulnerability"', color='b', size = 13)
ax2.set_ylabel('New Questions per Month Mentioning "Reentrancy" ', color='navy', size = 13)
ax2.grid(False)
ax.legend(['Security/Vulnerability'],loc = 2)

ax2.legend(['Reentrancy'],loc = 0)
plt.xlabel('January 2016 - August 2019', size = 13)
plt.title("Trends in Mentioning 'Reentrancy' or 'Security/Vulnerability'",size = 14)


plt.show()
```
